```python
tix = get_import_io_s_and_p_tickers()
df = get_collection_as_pandas_df(tix, 'stocks_test', update=False)
```

```python
test_date = datetime.datetime(2012,3,1,0,0)

# first find date in dataframe that is closest to chosen start date
closest = datetime.datetime(1900,1,1,0,0)
date_index = df.index
for ind, i in enumerate(df.index):
    #print "%s - %s - %s - %s" % (closest, i, math.fabs((i - start_date).days), math.fabs((closest - start_date).days))
    if math.fabs((i - test_date).days) < math.fabs((closest - test_date).days):
        closest = i
test_date = closest
```

```python
df_filt = df[df.index >= test_date - datetime.timedelta(days=90)]
df_filt = df_filt[df_filt.index <= test_date + datetime.timedelta(days=50)]

df_filt = df_filt.dropna(axis=1, how='any')
returns = (df_filt / df_filt.shift(1)).dropna(how='any')
returns_prod = returns.cumprod().dropna(how='any')
```

```python
# drop tickers with very extreme return values
extremes = returns.apply(lambda x: x.values > 1.25, axis=1).any()
df_filt = df_filt.drop(extremes.ix[extremes].index, 1)
returns = returns.drop(extremes.ix[extremes].index, 1)
returns_prod = returns_prod.drop(extremes.ix[extremes].index, 1)

extremes = returns.apply(lambda x: x.values < 0.75, axis=1).any()
df_filt = df_filt.drop(extremes.ix[extremes].index, 1)
returns = returns.drop(extremes.ix[extremes].index, 1)
returns_prod = returns_prod.drop(extremes.ix[extremes].index, 1)
```

```python
# get returns prod from test date
df_filt_by_test_date = df_filt[df_filt.index >= test_date]
df_filt_by_test_date = df_filt_by_test_date[df_filt_by_test_date.index <= test_date + datetime.timedelta(days=50)]

df_filt_by_test_date = df_filt_by_test_date.dropna(axis=1, how='any')
returns_by_test_date = (df_filt_by_test_date / df_filt_by_test_date.shift(1))#.dropna(how='any')
returns_prod_by_test_date = returns_by_test_date.cumprod()#.dropna(how='any')
returns_prod_by_test_date.ix[test_date,:] = 1.0
```

```python
fwd = {}
lengths = [1,3,5,7,9,11,13,15]
for i in lengths:
    fwd[i] = (returns_prod_by_test_date.ix[i]).dropna(how='any')
    fwd[i].sort()
```

```python
from p2 import *

tix = get_import_io_s_and_p_tickers()
df = get_collection_as_pandas_df(tix, 'stocks_test', update=False)
w = Window(df, start_date=(test_date-datetime.timedelta(days=90)), end_date=test_date, return_period_days=1)
pairs = w.pull_cointegrated_partners(date_strict=False)
```

```python
def get_cointegrated_list(ticker, pairs):
    # for a given ticker, return beta list containing only cointegrated stocks
    pairs_list = []
    for i in pairs:
        if ticker in i:
            pair = [t for t in i if t != ticker][0]  
            pairs_list.append(pair)
    return pairs_list
```

```python
def z_score_price_deviance_from_peers(pricing_data, ticker, pairs):
    
    if ticker not in pricing_data.columns:
        raise KeyError("Ticker not found in pricing data")
    
    # compare daily index of peers to dialy index of ticker by calculating the difference over each day
    # chooose the average last X days and determine the standard deviations outside normal range

    peers = get_cointegrated_list(ticker, pairs)
    
    checked_peers = []
    for p in peers:
        if p in pricing_data.columns:
            checked_peers.append(p)
    peers = checked_peers
    
    if len(peers) < 3:
        raise Exception("Insuficient number of pairs for this ticker")
    
    #diffs = self.returns[peers].sub(self.returns[ticker],axis=0) * -1.0
    diffs = pricing_data[peers].sub(pricing_data[ticker],axis=0) * -1.0

    z_scores = (diffs - diffs.mean()) / diffs.std()
    return z_scores.mean(axis=1)
    return z_scores.tail(10).mean(axis=0).mean()
```

```python
peer_comp = {}
for ind, i in enumerate(tix):
    try:
        series = z_score_price_deviance_from_peers(df_filt[df_filt.index < test_date], i, pairs)
        peer_comp[i] = series
    except:
        pass
peer_comp = pd.DataFrame.from_dict(peer_comp, orient='columns', dtype=None)
```

```python
ewma = pd.stats.moments.ewma
fig = plt.figure(figsize=(20, 4*len(lengths)))
gs = gridspec.GridSpec(len(lengths), 3)

myFmt = mdates.DateFormatter('%b-%d')

for ind, i in enumerate(lengths):
    
    tix = fwd[i].tail(30).index
    
    val_tix = []
    for t in tix:
        if t in peer_comp.columns:
            val_tix.append(t)
    tix = val_tix
    
    # individual peer comp difs z scores
    chart_data = peer_comp[tix]
    chart = plt.subplot(gs[ind, 0])
    chart.plot(chart_data.index, chart_data, alpha=0.2)
    chart.xaxis.set_major_formatter(myFmt)
    chart.set_title("Individual Peer Z-Scores - %s" % (i))

    # aggregated peer comp difs z scores
    avg = peer_comp[tix].mean(axis=1)
    emov = ewma(avg, com=5)
    avgs = pd.DataFrame.from_dict({'avg': avg, 'emov': emov}, orient='columns', dtype=None)
    
    chart_data = avgs
    chart = plt.subplot(gs[ind, 1])
    chart.plot(chart_data.index, chart_data, alpha=0.2)
    chart.xaxis.set_major_formatter(myFmt)
    chart.set_title("Aggregate Peer Z-Scores - %s" % (i))
    
    # returns product from test date to associated days fwd
    chart_data = returns_prod_by_test_date[tix]
    chart_data = chart_data[chart_data.index <= chart_data.index[i]]
    chart = plt.subplot(gs[ind, 2])
    chart.plot(chart_data.index, chart_data, alpha=0.2)
    chart.xaxis.set_major_formatter(myFmt)
    chart.set_title("Future Performance - %s" % (i))
```
