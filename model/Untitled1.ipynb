```python
# import libraries
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

import pandas.io.data as web
import datetime

from p3 import *
```

```python
portfolio_returns = pd.io.parsers.read_csv('model_output/test_results.csv', index_col=0, parse_dates=True)

date_range = portfolio_returns.index[portfolio_returns.index > datetime.datetime(2009,12,31,0,0)]
#date_range = portfolio_returns.index

portfolio_returns = portfolio_returns.ix[date_range]

index_tix = ['^GSPC', '^IXIC']
index = get_collection_as_pandas_df(index_tix, 'index_test')

index = index.ix[date_range][['^GSPC',]]


index_returns = (index / index.shift(1))
portfolio_returns['index'] = index_returns

portfolio_returns_prod = portfolio_returns.cumprod()
portfolio_returns = portfolio_returns.dropna(how='any')
portfolio_returns_prod = portfolio_returns_prod.dropna(how='any')
```

```python
p = index.head(1)['^GSPC'].values[0]
# build list of change in prices over time
ps = [p,]
for r in portfolio_returns['index'].values:
    ps.append(r*ps[-1])
price_build_returns = pd.Series(ps, index=index['^GSPC'].index)
index['price_build_returns'] = price_build_returns

# build list of change in prices over time
ps = [p,]
for r in portfolio_returns_prod['index'].values:
    ps.append(r*p)
price_build_returns_prod = pd.Series(ps, index=index['^GSPC'].index)
index['price_build_returns_prod'] = price_build_returns_prod

index.plot()
```
